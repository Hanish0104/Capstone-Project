```python
from datasets import load_dataset

ds = load_dataset("AISE-TUDelft/PY150k")
train_data = ds["train"]
test_data = ds["test"]

examples = test_data.select(range(10))
```

```python
df_train = train_data.to_pandas()
df_test = test_data.to_pandas()
df_train.head()
```

```python
model_name = "bigcode/starcoder2-7b"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)
```

```python
inputs = examples["input"]
references = examples["full_line"]

predictions = []

for i, code in enumerate(inputs):
    if code and code.strip():
        input_ids = tokenizer(code, return_tensors="pt").input_ids.to(model.device)

        if input_ids.shape[1] > 0:
            output_ids = model.generate(input_ids, max_new_tokens=128)
            pred = tokenizer.decode(output_ids[0], skip_special_tokens=True)
            predictions.append(pred)
        else:
            predictions.append(f"[EMPTY_TOKENIZED_INPUT_FOR_SAMPLE_{i}]")
    else:
        predictions.append(f"[EMPTY_INPUT_FOR_SAMPLE_{i}]")

if predictions:
    print("Input Code:\n", inputs[0])
    print("\nReference Docstring:\n", references[0])
    print("\nPredicted Docstring:\n", predictions[0])
else:
    print("No predictions were generated (all inputs might have been empty).")
```

```python
import sacrebleu

bleu = sacrebleu.corpus_bleu(predictions, [references])
print("BLEU Score:", bleu.score)
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = [scorer.score(ref, pred) for ref, pred in zip(references, predictions)]

avg_scores = {}
for key in scores[0].keys():
    avg_scores[key] = sum([s[key].fmeasure for s in scores]) / len(scores)

print("Average ROUGE Scores:", avg_scores)
```
